**Classify Movies Based on Ratings [Good,Better and Best]**

```python
df.columns
```

```python
def rating(rating):
  if rating >= 7.0:
    return 'Excellent'
  elif rating >= 6.0:
    return 'Good'
  else:
    return 'Better'
```

```python
df['rating_cat'] = df['Rating'].apply(rating)
```
